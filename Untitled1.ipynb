```python
import numpy as np

import matplotlib.pyplot as plt

from scipy.linalg import lstsq

# Generate a sparse sine wave signal

signal_length = 1000

time = np.linspace(0, 1, signal_length)

sparse_signal = np.sin(2 * np.pi * 5 * time) + np.sin(2 * np.pi * 20 * time)

sparse_signal[np.random.choice(range(signal_length), int(signal_length * 0.9), replace=False)] = 0

# Define the measurement matrix

measurement_matrix = np.random.randn(100, signal_length)

# Take random measurements of the sparse signal

measurements = np.dot(measurement_matrix, sparse_signal)

# Perform sparse signal recovery using L1-norm minimization

recovered_signal = lstsq(measurement_matrix, measurements, cond=None)[0]

# Plot the original and recovered signals

plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)

plt.plot(time, sparse_signal)

plt.title('Original Sparse Sine Wave Signal')

plt.subplot(2, 1, 2)

plt.plot(time, recovered_signal)

plt.title('Recovered Sine Wave Signal')

plt.tight_layout()

plt.show()

 
```
