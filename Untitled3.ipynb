```python
from scipy.optimize import minimize

import numpy as np

import matplotlib.pyplot as plt

 

# Objective function

def objective(x):

   cost_coefficient = 10

   return -cost_coefficient * x

 

# Initial guess

x0 = 50

 

# Solve the optimization problem

solution = minimize(objective, x0)

 

# Optimal power generation value

power_opt = solution.x[0]

 

# Plotting the objective function

power = np.linspace(0, 100, 100)

cost = -objective(power)

 

# Plotting the cost curve

plt.plot(power, cost)

plt.xlabel('Power Generation')

plt.ylabel('Cost')

plt.title('Unconstrained Optimization')

plt.scatter(power_opt, -solution.fun, color='red', label='Optimal Solution')

plt.legend()

plt.show()

 

# Print the optimal solution

print('Optimal Power Generation:', power_opt)

print('Minimum Cost:', -solution.fun)

```
